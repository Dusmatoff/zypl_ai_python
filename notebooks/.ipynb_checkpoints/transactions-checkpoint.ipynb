```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV data into a DataFrame
df = pd.read_csv('transactions.csv', low_memory=False)
```

```python
df['date'] = pd.to_datetime(df['date']).dt.date
```

```python
# Display the first few rows of the DataFrame to understand its structure
df.head()
```

```python
df['amount'].describe()
```

```python
# Total amount of transactions
total_amount = df['amount'].sum()
total_amount
```

```python
# Total number of transactions
total_transactions = df['amount'].count()
total_transactions
```

```python
# Total amount of transactions by status
amount_by_status = df.groupby('status')['amount'].sum()
amount_by_status
```

```python
# Total amount of transactions by user
amount_by_user = df.groupby('user_name')['amount'].sum()
amount_by_user
```

```python
# Plot total amount of transactions by user (Top 10 Users)
top_10_users = amount_by_user.nlargest(10)
plt.figure(figsize=(12, 8))
top_10_users.plot(kind='bar', color='lightgreen')
plt.title('Total Amount of Transactions by Top 10 Users')
plt.xlabel('User')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Plot total amount of transactions by status
plt.figure(figsize=(10, 6))
amount_by_status.plot(kind='bar', color='skyblue')
plt.title('Total Amount of Transactions by Status')
plt.xlabel('Status')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()
```
