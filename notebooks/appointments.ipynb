```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('appointments.csv') # Dataframe, read from file
```

```python
df.describe() #main describe of data
```

```python
la_events = df[df['city_name'] == 'Los Angeles'] # Filter appointments by city
la_events
```

```python
status_counts = df['status'].value_counts() # Count of appointments by status
status_counts
```

```python
city_counts = df['city_name'].value_counts() # Count of appointments by city
city_counts
```

```python
# Group by hall name and count the number of events in each hall
events_by_hall = df.groupby('hall_name').size()
events_by_hall
```

```python
# Create a pivot table with city_name as rows, status as columns, and count of events as values
pivot_table = pd.pivot_table(df, values='id', index='city_name', columns='status', aggfunc='count')
pivot_table
```

```python
# Convert the 'date' column to datetime if it's not already
df['date'] = pd.to_datetime(df['date'])

# Filter the data to include only dates in January 2024
january_2024_df = df[(df['date'].dt.year == 2024) & (df['date'].dt.month == 1)]

# Group the filtered data by date to count the number of events each day
events_per_day_january_2024 = january_2024_df.groupby('date').size()

# Plot a line chart for the number of events per day in January 2024
plt.figure(figsize=(10, 5))
plt.plot(events_per_day_january_2024.index, events_per_day_january_2024.values, marker='o', color='b')
plt.title('Number of Events Per Day (January 2024)')
plt.xlabel('Date')
plt.ylabel('Number of Events')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart of event statuses
plt.figure(figsize=(7, 7))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'coral'])
plt.title('Distribution of Event Status')
plt.show()
```

```python
# Create subplots: bar chart of events by status and pie chart of events by city
fig, ax = plt.subplots(1, 2, figsize=(15, 7))

# Bar chart of events by status
ax[0].bar(status_counts.index, status_counts.values, color='skyblue')
ax[0].set_title('Count of Events by Status')
ax[0].set_xlabel('Status')
ax[0].set_ylabel('Number of Events')

# Pie chart of events by city
ax[1].pie(city_counts, labels=city_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue', 'lightgreen'])
ax[1].set_title('Distribution of Events by City')

plt.tight_layout()
plt.show()
```
